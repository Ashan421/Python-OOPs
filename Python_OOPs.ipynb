1. What is Object-Oriented Programming (OOP) ?

 - OOPs refers to languages that use objects in programming. Object-oriented programming aims to implement real-world entities like inheritance, hiding, polymorphism, etc in programming. The main aim of OOP is to bind together the data and the functions that operate on them so that no other part of the code can access this data except that function.
    -  Class: A class is a user-defined data type. It consists of data members and member functions, which can be accessed and used by creating an instance of that class. It represents the set of properties or methods that are common to all objects of one type. A class is like a blueprint for an object.
    - Object: Object-Oriented Programming and represents the real-life entities. An Object is an instance of a Class. When a class is defined, no memory is allocated but when it is instantiated  memory is allocated. An object has an identity, state, and behavior. Each object contains data and code to manipulate the data. Objects can interact without having to know details of each other's data or code, it is sufficient to know the type of message accepted and type of response returned by the objects.
    - Data Abstraction: One of the most essential and important features of object-oriented programming. Data abstraction refers to providing only essential information about the data to the outside world, hiding the background details or implementation.This is what abstraction is.
    -  Encapsulation: Defined as the wrapping up of data under a single unit. It is the mechanism that binds together code and the data it manipulates. In Encapsulation, the variables or data of a class are hidden from any other class and can be accessed only through any member function of their class in which they are declared. As in encapsulation, the data in a class is hidden from other classes, so it is also known as data-hiding.
    - Inheritance: Inheritance is an important pillar of OOP. The capability of a class to derive properties and characteristics from another class is called Inheritance. When we write a class, we inherit properties from other classes. So when we create a class, we do not need to write all the properties and functions again and again, as these can be inherited from another class that possesses it. Inheritance allows the user to reuse the code whenever possible and reduce its redundancy.
    - Polymorphism: Having many forms,In simple words, we can define polymorphism as the ability of a message to be displayed in more than one form. This is called polymorphism.
    - Dynamic Binding: The code to be executed in response to the function call is decided at runtime. Dynamic binding means that the code associated with a given procedure call is not known until the time of the call at run time. Dynamic Method Binding One of the main advantages of inheritance is that some derived class D has all the members of its base class B. Once D is not hiding any of the public members of B, then an object of D can represent B in any context where a B could be used. This feature is known as subtype polymorphism.
    - Message Passing: Communication used in object-oriented programming as well as parallel programming. Objects communicate with one another by sending and receiving information to each other. A message for an object is a request for execution of a procedure and therefore will invoke a function in the receiving object that generates the desired results. Message passing involves specifying the name of the object, the name of the function, and the information to be sent.

4.  What is the difference between abstraction and encapsulation?

 - Abstraction and encapsulation are two fundamental concepts in Object-Oriented Programming that often get confused because they both involve "hiding" details. However, they serve different purposes and operate at different levels.
 - Abstraction: "What" an object does.
     - Abstraction focuses on hiding complex implementation details and showing only the essential features or functionalities that are relevant to the user. It's about providing a simplified, high-level view of an object or system.
     - Purpose: To reduce complexity, make systems easier to understand and use, and allow users to interact with objects without needing to know "how" they work internally.
     - Focus: "What" an object does or "what" functionality it provides.
     - Level: Design or interface level. We define the public interface  that users will interact with.
 - Encapsulation: "How" an object protects its data
     - Encapsulation is the bundling of data (attributes) and the methods that operate on that data into a single unit (a class), and restricting direct access to the data from outside the unit. It's about data hiding and protecting an object's internal state.
     - Purpose: To protect the integrity of an object's data, prevent unauthorized access or modification, and ensure that data is manipulated only through defined methods. This leads to better data security and maintainability.
     - Focus: "How" an object manages its internal data and "how" that data can be accessed or modified.
     - Level: Implementation level.We control the visibility of members within a class.

7.  What is polymorphism in OOP?

 - Polymorphism is a fundamental concept in Object-Oriented Programming \that allows objects of different types to be accessed through the same interface. This means that a single function or method can operate on different types of objects, each providing its own implementation of the interface.

     - Compile-Time Polymorphism : Compile-time polymorphism, also known as static polymorphism, is achieved through method overloading or operator overloading. In method overloading, multiple methods with the same name but different parameters are defined within the same class. The compiler determines which method to call based on the method signature.
     - Runtime Polymorphism : Runtime polymorphism, also known as dynamic polymorphism, is achieved through method overriding. In method overriding, a subclass provides a specific implementation of a method that is already defined in its superclass. The method to be executed is determined at runtime based on the object's actual type.

10. What are class and static methods in Python?

  - In Python, class methods and static methods are two types of methods that belong to a class rather than an instance of the class. They are defined differently and serve distinct purposes:

    - Class Methods : A class method is a method that operates on the class itself rather than an instance of the class. It has access to the class and can modify its state.
      - How to Define: Use the @classmethod decorator.
      - First Parameter: The first parameter is cls, which refers to the class.
      - Use Cases:
       Factory methods that create instances in a specific way.
       Accessing or modifying class-level attributes.
    - Example:
          class MyClass:
          class_variable = "Hello, Class!"

          @classmethod
          def class_method(cls):
          return f"Class variable is: {cls.class_variable}"

          # Usage
          print(MyClass.class_method())      

    - Static Methods: A static method is a method that does not operate on the class or its instances. It behaves like a regular function but is logically grouped within the class.
      - How to Define: Use the @staticmethod decorator.
      - First Parameter: It does not take self or cls as its first parameter.
      - Use Cases:
       Utility functions that are related to the class but do not need access to class or instance data.
    - Example:

            class MyClass:
            @staticmethod
            def static_method(x, y):
            return x + y

            # Usage
            print(MyClass.static_method(5, 3))  
           

11. What is method overloading in Python?
 - Method overloading in Python refers to the ability to define multiple methods in the same class with the same name but different parameters.Unlike statically-typed languages like Java or C++, Python does not support traditional method overloading directly. Instead, Python achieves similar functionality through default arguments, variable-length arguments, or type checking within a single method.
     - Dynamic Typing: Python is a dynamically-typed language. When we define a function or method,we don't explicitly declare the types of its parameters.
     - Last Definition Wins: If we define multiple methods with the same  name in a Python class, the last definition will always override the previous ones. Python treats methods as attributes, and assigning a new method to the same name simply overwrites the old one.

    - Using Default Arguments : Define a single method with default parameter values to handle different cases.
          class Calculator:

          calc = Calculator()

    - Using Variable-Length Arguments : Use *args to accept a variable number of arguments.
          class Calculator:

          return sum(args)

          calc = Calculator()
          print(calc.add(5))         
          print(calc.add(5, 10))

    - Using Type Checking : Differentiate behavior based on the types of arguments passed.

12. What is method overriding in OOP?
  - Method overriding in Object-Oriented Programming (OOP) in Python refers to the process where a subclass provides a specific implementation of a method that is already defined in its parent class. The overridden method in the subclass has the same name, parameters, and return type as the method in the parent class. This allows the subclass to customize or completely replace the behavior of the parent class's method.

  - Key Points:
     - Inheritance: Method overriding can only occur in the context of inheritance. There must be a parent-child relationship between classes.
     
     - Same Method Signature: The overriding method in the subclass must have the exact same name, number of parameters, and parameter types  as the method in the superclass. This is often referred to as having the "same signature."

     - Runtime Polymorphism: When an overridden method is called on an object, the version of the method that gets executed is determined at runtime based on the actual type of the object, not the type of the reference variable. This is why method overriding is crucial for achieving runtime polymorphism.

14. Why is polymorphism important in OOP?
 - Polymorphism is incredibly important in Object-Oriented Programming because it brings significant benefits that lead to more flexible, maintainable, and extensible software systems. Here's a breakdown of its key importance:

    - Increased Flexibility and Adaptability:Generic Code:
        - Polymorphism allows me to write generic code that can operate on objects of different, but related, types. We can define a common interface, and then objects of various subclasses can be processed using that same interface.
        - Example: Imagine a drawing application. With polymorphism,I can have a list of Shape objects. I can then iterate through this list and call a draw() method on each Shape, and the correct draw() method for each specific shape will be executed automatically. We don't need if/else statements to check if shape is a Circle, draw it this way; if shape is a Square, draw it that way.

    - Code Reusability:  
        - By defining common behaviors in a superclass and allowing subclasses to override them, we avoid duplicating code. The generic behavior is defined once, and specific variations are handled within the subclasses.
        - This makes our codebase leaner and more efficient.

   - Improved Code Readability and Maintainability:
        - Polymorphism makes code cleaner and easier to understand because it focuses on what an object does rather than what its exact type is.
        - When we need to modify or extend the behavior, we typically only need to make changes in the relevant subclass's overridden method, without affecting other parts of the system that interact with the common interface. This significantly reduces maintenance effort.

  - Enhanced Extensibility:  
        - Polymorphism supports the Open/Closed Principle, a key SOLID principle: "Software entities should be open for extension, but closed for modification."
        - We can easily add new subclasses that implement the common interface without having to modify existing code. For example, in the drawing application, if we introduce a new Hexagon shape, i just create a Hexagon class, override its draw() method, and our existing drawing loop will automatically work with Hexagon objects without any changes. This is incredibly powerful for scaling applications.

 - Decoupling and Loose Coupling:
      - Polymorphism helps to reduce dependencies between different parts of our code. our client code  interacts with a high-level abstraction, rather than being tightly coupled to specific concrete subclasses.

      - This loose coupling makes our system more robust to changes. If the internal implementation of a subclass changes, the client code doesn't need to be updated.

   - Facilitates Design Patterns:
      - Many well-known design patterns  heavily rely on polymorphism to achieve their flexibility and reusability goals. These patterns provide proven solutions to common software design problems, and polymorphism is a cornerstone for implementing them effectively.

  - Easier Testing:
      - Polymorphism can make unit testing easier. We can test a component by substituting its dependencies with mock objects that adhere to the same polymorphic interface, allowing me to isolate and test specific units of code effectively.

15. What is an abstract class in Python?
 - An abstract class in Python is a class that cannot be instantiated directly. Its primary purpose is to define a common interface that its subclasses must implement. Abstract classes often contain one or more abstract methods, which are methods declared without an implementation.
 - Key Characteristics of Abstract Classes:
      - Cannot Be Instantiated: We cannot create an object directly from an abstract class. Attempting to do so will result in a TypeError.
      - Requires Subclassing: Abstract classes are designed to be inherited from. Their value lies in forcing subclasses to adhere to a specific structure and implement certain behaviors.
      - Define Common Interface: They define a contract. Any concrete  subclass of an abstract class must provide implementations for all of its abstract methods. If a subclass fails to implement an abstract method, it also becomes an abstract class and cannot be instantiated.
      - Can Have Concrete Methods and Attributes: While they must have at least one abstract method , abstract classes can also have regular methods with implementations and regular attributes. These can be inherited and used directly by subclasses.

            from abc import ABC, abstractmethod
            class Polygon(ABC):
            @abstractmethod
            def no_of_sides(self):
            pass

            class Triangle(Polygon):
            def no_of_sides(self):
            print("I have 3 sides")

            class Quadrilateral(Polygon):
            def no_of_sides(self):
            print("I have 4 sides")

    - In this example, Polygon is an abstract class with an abstract method no_of_sides. The Triangle and Quadrilateral classes inherit from Polygon and provide their own implementations of the no_of_sides method.

17. What is the difference between a class variable and an instance variable?
  - Class variables and instance variables are two types of variables that differ in their scope, behavior, and usage.
  - Class Variable
      - Definition: A variable that is shared across all instances of a class. It is defined within the class but outside any instance methods.
      - Scope: Belongs to the class itself and is shared by all instances of the class.
      - Usage: Used for data that should be the same for all instances, like constants or shared state.
      - Modification: Changing a class variable affects all instances unless explicitly overridden in an instance.

  -  Instance Variable
      - Definition: A variable that is unique to each instance of a class. It is defined within instance methods using self.
      - Scope: Belongs to the specific instance of the class.
      - Usage: Used for data that is unique to each instance.
      - Modification: Changing an instance variable affects only that specific instance.

18. What is multiple inheritance in Python?
  - Python, multiple inheritance is a feature that allows a class to inherit attributes and methods from more than one parent class. This means a single child class can derive functionality from multiple base classes, enabling code reuse and flexibility.
  - Example of Multiple Inheritance:

          class Parent1:
          def greet(self):
          print("Hello from Parent1!")

          class Parent2:
          def welcome(self):
          print("Welcome from Parent2!")

          pass

          # Creating an object of the Child class
          child = Child()

       - Method Resolution Order (MRO): Python uses the C3 Linearization algorithm to determine the order in which methods are inherited when there are conflicts. We can check the MRO using ClassName.mro() or help(ClassName).

     - Potential Pitfalls: While powerful, multiple inheritance can lead to complexity, such as:

     - Ambiguity: If two parent classes have methods with the same name, it can be unclear which method will be called.
     - Diamond Problem: When two parent classes inherit from a common base class, and the child class inherits from both, it can create confusion about which path to follow.
  - Example of the Diamond Problem:
          class A:

          pass

19. Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in Python?

 - In Python, the __repr__ and __str__ methods are used to define string representations of objects. The __repr__ method is intended for developers, providing an unambiguous representation of the object, while __str__ is for users, offering a readable description.
 - Example :

           class Person:
           def __init__(self, first_name, last_name, age):
           self.first_name = first_name
           self.last_name = last_name
           self.age = age

           def __repr__(self):
           return f'Person("{self.first_name}", "{self.last_name}", {self.age})'

           def __str__(self):
           return f'{self.first_name} {self.last_name}, Age: {self.age}'

           person = Person('John', 'Doe', 30)
           print(repr(person))
           print(str(person))

  - Key Differences :

      - Purpose: __repr__: Provides a detailed and unambiguous string representation for developers. It should ideally be a valid Python expression that can recreate the object. __str__: Offers a readable and user-friendly string representation.

      - Usage: repr(): Calls the __repr__ method. str(): Calls the __str__ method. If __str__ is not defined, it falls back to __repr__.

  - Example with Built-in Class :
       
          import datetime

          now = datetime.datetime.now()
          print(repr(now))
          print(str(now))

  - In this example, repr(now) provides a detailed representation that can recreate the object, while str(now) offers a more readable format.

20. What is the significance of the ‘super()’ function in Python?

 - The super() function in Python is a powerful and essential tool, especially in object-oriented programming. It plays a significant role in enabling efficient and clean inheritance.

   1. Access Parent Class Methods : The super() function allows us to call methods from a parent in a child without explicitly naming the parent class. This is particularly useful when overriding methods in the child class but still needing to retain and extend the functionality of the parent class.

            class Parent:
            def greet(self):
            print("Hello from Parent!")

            class Child(Parent):
            def greet(self):
            super().greet()  # Calls the parent class's greet method
            print("Hello from Child!")

            obj = Child()
            obj.greet()

   2. Support for Multiple Inheritance : In Python, super() is especially helpful in multiple inheritance scenarios. It ensures that the Method Resolution Order is followed, which determines the order in which classes are searched for a method. This avoids redundancy and ensures that each class in the hierarchy is called only once.

          class A:
          def show(self):
          print("A")

          class B(A):
          def show(self):
          print("B")
          super().show()

          class C(A):
          def show(self):
          print("C")
          super().show()

           class D(B, C):
           def show(self):
           print("D")
           super().show()

           obj = D()
           obj.show()

   3. Improves Code Maintainability : By using super(), you avoid hardcoding the parent class name. This makes your code more maintainable and flexible, as changes to the class hierarchy won't require you to update method calls manually.
            class Parent:
            def __init__(self, name):

 - @staticmethod : A @staticmethod is a method that does not receive any implicit first argument. It behaves like a regular function but belongs to the class's namespace.             class Date:

            is_valid = Date.is_date_valid('01-01-2025')
            print(is_valid)

   - In this example, is_date_valid is a static method that validates a date string without needing any information about the class or its instances.

  - @classmethod : A @classmethod is a method that receives the class itself as the first argument, typically named cls. It can access and modify class state that applies across all instances of the class. It is often used for factory methods that create instances of the class using different sets of parameters.
      - Example:   

             class Date:
             def __init__(self, day=0, month=0, year=0):
             self.day = day
             self.month = month
             self.year = year

             @classmethod
             def from_string(cls, date_as_string):
             day, month, year = map(int, date_as_string.split('-'))
